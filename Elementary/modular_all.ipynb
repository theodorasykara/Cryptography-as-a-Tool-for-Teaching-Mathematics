```python
# Διαδραστικό πρόγραμμα για να κατανοήσουμε την έννοια του "υπολοίπου".
# Περιλαβάνει την περιπτωση του αδύνατου αποτελέσματος και την πράξη της διάιρεσης
def euclidean_modular_arithmetic(operation, modulus, alpha, beta):
    """
    Διαδραστικό πρόγραμμα για την κατανόηση του υπολοίπου και της modular αριθμητικής.
    operation: Η πράξη που θα εκτελεστεί (πρόσθεση, αφαίρεση, πολλαπλασιασμός, διαίρεση).
    modulus: Το μέτρο (modulus) για την modular αριθμητική.
    alpha: Ο πρώτος αριθμός.
    beta: Ο δεύτερος αριθμός.
    """
    output = Output()

    # Υπολογισμός του σωστού αποτελέσματος
    if operation == "Πρόσθεση (Addition)":
        correct_result = (alpha + beta) % modulus
        operation_symbol = "+"
        detailed_steps = (
            f"({alpha} + {beta}) mod {modulus} = {alpha + beta} mod {modulus} = {correct_result}"
        )
    elif operation == "Αφαίρεση (Subtraction)":
        correct_result = (alpha - beta) % modulus
        operation_symbol = "-"
        detailed_steps = (
            f"({alpha} - {beta}) mod {modulus} = {alpha - beta} mod {modulus} = {correct_result}"
        )
    elif operation == "Πολλαπλασιασμός (Multiplication)":
        correct_result = (alpha * beta) % modulus
        operation_symbol = "×"
        detailed_steps = (
            f"({alpha} × {beta}) mod {modulus} = {alpha * beta} mod {modulus} = {correct_result}"
        )
    elif operation == "Διαίρεση (Division)":
        try:
            beta_inverse = Integer(beta).inverse_mod(modulus)
            correct_result = (alpha * beta_inverse) % modulus
            operation_symbol = "÷"
            detailed_steps = (
                f"({alpha} ÷ {beta}) mod {modulus} = ({alpha} × {beta_inverse}) mod {modulus} = {correct_result}"
            )
        except ZeroDivisionError:
            correct_result = None
            operation_symbol = "÷"
            detailed_steps = (
                f"({alpha} ÷ {beta}) mod {modulus} είναι αδύνατο γιατί το {beta} "
                f"δεν έχει αντίστροφο modulo {modulus}. Βεβαιωθείτε ότι {beta} "
                f"και {modulus} είναι πρώτοι μεταξύ τους."
            )

    # Εμφάνιση ερώτησης στον μαθητή
    print(f"\nΥπολογίστε: ({alpha} {operation_symbol} {beta}) mod {modulus}")
    
    # Δημιουργία πεδίου εισαγωγής για την απάντηση
    student_guess_input = widgets.BoundedIntText(
        value=0, min=0, max=modulus-1, description="Απάντηση:"
    )
    submit_button = widgets.Button(description="Υποβολή")
    
    # Συνάρτηση που εκτελείται όταν ο χρήστης πατήσει το κουμπί υποβολής
    def check_answer(b):
        with output:
            output.clear_output()
            student_guess = student_guess_input.value
            
            if correct_result is None:
                print("\n⚠️ Η πράξη είναι αδύνατη.")
                print(detailed_steps)
                print("\nΠροσπαθήστε με έναν διαφορετικό αριθμό για τον διαιρέτη (β).")
            else:
                if student_guess == correct_result:
                    print("\n✅ Συγχαρητήρια! Η απάντησή σας είναι σωστή!")
                else:
                    print("\n❌ Η απάντησή σας δεν είναι σωστή.")
                    print(f"Το σωστό αποτέλεσμα είναι: {correct_result}.")
                
                # Αναλυτικό σκεπτικό για την πράξη
                print("\n🔍 Αναλυτική διαδικασία:")
                print(detailed_steps)

                # Ειδικές συμβουλές για κάθε πράξη
                print("\n--- Χρήσιμες Συμβουλές ---")
                if operation == "Πρόσθεση (Addition)":
                    print("- Προσθέστε τους αριθμούς και βρείτε το υπόλοιπο της διαίρεσης με το n.")
                elif operation == "Αφαίρεση (Subtraction)":
                    print(
                        "- Υπολογίστε τη διαφορά των αριθμών και φέρτε το αποτέλεσμα εντός του διαστήματος [0, n)."
                    )
                elif operation == "Πολλαπλασιασμός (Multiplication)":
                    print("- Υπολογίστε το γινόμενο και βρείτε το υπόλοιπο της διαίρεσης με το n.")
                elif operation == "Διαίρεση (Division)":
                    print(
                        "- Βρείτε το αντίστροφο του διαιρέτη (αν υπάρχει) και πολλαπλασιάστε το με τον πρώτο αριθμό."
                    )
    
    # Σύνδεση κουμπιού με τη συνάρτηση ελέγχου
    submit_button.on_click(check_answer)
    
    # Εμφάνιση widgets
    display(student_guess_input, submit_button, output)

# Διαδραστικό πρόγραμμα
interact(
    euclidean_modular_arithmetic,
    operation=[
        "Πρόσθεση (Addition)",
        "Αφαίρεση (Subtraction)",
        "Πολλαπλασιασμός (Multiplication)",
        "Διαίρεση (Division)",
    ],
    modulus=(2, 12, 1),
    alpha=(0, 100, 1),
    beta=(0, 100, 1)
)

```
