```python
# Διαδραστική Δραστηριότητα 1 
# Έλεγχος για το αν ένας αριθμός είναι πρώτος
# Ο μαθητής επιλέγει αριθμό και δίνει απάντηση
# Το πρόγραμμα ελέγχει αν είναι σωστή και δίνει ανατροφοδότηση
@interact
def ελεγχος_πρώτων(
    n1 = selector([17, 18, 25, 37], label="Επίλεξε αριθμό:", buttons=True),
    απάντηση = selector(["-- Επιλογή --", "Είναι πρώτος", "Δεν είναι πρώτος"], label="Η απάντησή σου:")
):
    # Αν ο μαθητής δεν έχει απαντήσει ακόμα, δεν δίνουμε feedback
    if απάντηση == "-- Επιλογή --":
        print("Διάλεξε απάντηση για να δεις αν είναι σωστή.")
        return

    σωστό = is_prime(n1)
    ειναι = "Είναι πρώτος" if σωστό else "Δεν είναι πρώτος"

    print(f"Ο αριθμός {n1} {'' if σωστό else 'δεν '}είναι πρώτος.")

    if απάντηση == ειναι:
        print("Μπράβο! Η απάντησή σου είναι σωστή.")
    else:
        print("Η απάντησή σου είναι λάθος. Θυμήσου: πρώτος αριθμός έχει μόνο δύο διαιρέτες, το 1 και τον εαυτό του.")
```

```python
# Διαδραστική Δραστηριότητα 1 
# Υπολογισμός πρώτων αριθμών μέχρι ένα επιλεγμένο όριο
# Ο μαθητής βλέπει τους πρώτους και βγάζει μαθηματικά συμπεράσματα
@interact
def πρωτοι_μεχρι_οριο(
    όριο=slider(10, 100, 1, default=30, label="Διάλεξε άνω όριο:"),
    δείξε=checkbox(False, " Εμφάνισε τους πρώτους αριθμούς έως το όριο")
):
    if not δείξε:
        print(f" Διάλεξες όριο: {όριο}. Τώρα επίλεξε το κουτάκι για να δεις τους πρώτους αριθμούς μέχρι το {όριο}.")
        return

    primes = list(prime_range(2, όριο + 1))

    print(f" Υπολογισμός πρώτων αριθμών από 2 έως {όριο}...")
    print(f" Πρώτοι αριθμοί: {primes}")
    print(f" Πλήθος πρώτων: {len(primes)}")
    print(f" Ο μέγιστος πρώτος στο εύρος είναι: {max(primes) if primes else '–'}")

    # Προαιρετικό μαθηματικό tip
    if όριο < 20:
        print("Παρατήρησε πόσοι λίγοι πρώτοι υπάρχουν σε μικρούς αριθμούς.")
    elif όριο > 50:
        print("Οι πρώτοι αραιώνουν όσο μεγαλώνει το όριο – αλλά δεν σταματούν ποτέ!")


```

```python
# Διαδραστική Δραστηριότητα 1 
# Ο μαθητής προσπαθεί να μαντέψει τον επόμενο πρώτο αριθμό μετά από έναν ακέραιο
# Το πρόγραμμα ελέγχει πόσο κοντά είναι και δίνει ενισχυτική ανατροφοδότηση

@interact
def επόμενος_πρώτος(
    x = slider(90, 120, 1, default=100, label="Επέλεξε έναν αριθμό:"),
    υπόθεση = input_box(default=0, label="Ποιος πιστεύεις ότι είναι ο επόμενος πρώτος αριθμός;"),
    εμφάνιση = checkbox(False, " Έλεγχος απάντησης")
):
    if not εμφάνιση:
        print(f" Επέλεξες το {x}. Πληκτρολόγησε την υπόθεσή σου και μετά επίλεξε το κουτάκι για να δεις αν έχεις δίκιο.")
        return

    επόμενος = next_prime(x)

    print(f"Ο επόμενος πρώτος μετά το {x} είναι:  {επόμενος}")

    # Έλεγχος ορθότητας
    if υπόθεση == επόμενος:
        print("Σωστά! Πολύ καλά, έχεις καταλάβει πώς κατανέμονται οι πρώτοι αριθμοί.")
    else:
        print(f"Όχι ακριβώς. Εσύ είπες: {υπόθεση}, αλλά η σωστή απάντηση είναι: {επόμενος}.")
        διαφορά = abs(υπόθεση - επόμενος)
        if διαφορά == 1:
            print(" Ήσουν πολύ κοντά!")
        elif διαφορά <= 5:
            print("Καλή προσπάθεια!")
        else:
            print("Δες ξανά πώς κατανέμονται οι πρώτοι αριθμοί.")
    
    print("Θυμήσου: Οι πρώτοι αριθμοί γίνονται πιο σπάνιοι όσο μεγαλώνουν οι ακέραιοι.")

```

```python
# Διαδραστική Δραστηριότητα 1 
# Ο μαθητής υπολογίζει πόσοι πρώτοι υπάρχουν έως το 30 και το 100
# Δίνει πρόβλεψη για το πώς αλλάζει η συχνότητά τους με το όριο
@interact
def δραστηριότητα_πρώτων(
    απάντηση_30 = input_box(default=0, label="Πόσους πρώτους αριθμούς πιστεύεις ότι υπάρχουν έως το 30;"),
    απάντηση_100 = input_box(default=0, label="Πόσους πρώτους αριθμούς υπάρχουν έως το 100;"),
    πρόβλεψη = selector([
        "Η συχνότητα των πρώτων παραμένει σταθερή",
        "Οι πρώτοι αραιώνουν όσο μεγαλώνει το όριο",
        "Οι πρώτοι γίνονται συχνότεροι όσο μεγαλώνει το όριο"
    ], label="Τι παρατηρείς για τη συχνότητα των πρώτων καθώς μεγαλώνει το όριο;"),
    εμφάνιση = checkbox(False, "Έλεγχος απαντήσεων")
):
    # Σωστές απαντήσεις
    σωστό_30 = len(prime_range(2, 31))    # 10 πρώτοι
    σωστό_100 = len(prime_range(2, 101))  # 25 πρώτοι

    if not εμφάνιση:
        print("Συμπλήρωσε τις απαντήσεις σου και μετά επίλεξε το κουτάκι για να δεις αν είναι σωστές.")
        return

    # Ανατροφοδότηση για αριθμητικά
    print(f"Απάντησες ότι υπάρχουν {απάντηση_30} πρώτοι έως το 30.")
    if απάντηση_30 == σωστό_30:
        print("✓ Σωστά για το 30!")
    else:
        print(f"✗ Λάθος για το 30. Η σωστή απάντηση είναι {σωστό_30}.")

    print(f"\nΑπάντησες ότι υπάρχουν {απάντηση_100} πρώτοι έως το 100.")
    if απάντηση_100 == σωστό_100:
        print("✓ Σωστά για το 100!")
    else:
        print(f"✗ Λάθος για το 100. Η σωστή απάντηση είναι {σωστό_100}.")

    # Ανατροφοδότηση για παρατήρηση
    print(f"\nΠρόβλεψή σου: «{πρόβλεψη}»")
    if πρόβλεψη == "Οι πρώτοι αραιώνουν όσο μεγαλώνει το όριο":
        print("✓ Σωστά! Οι πρώτοι εμφανίζονται λιγότερο συχνά όσο μεγαλώνουν οι αριθμοί.")
    else:
        print("✗ Η σωστή παρατήρηση είναι ότι οι πρώτοι αραιώνουν όσο μεγαλώνει το όριο.")

    # Μαθηματικό σχόλιο
    αναλογία_30 = float(σωστό_30 / 30)
    αναλογία_100 = float(σωστό_100 / 100)

    print(f"\nΠυκνότητα πρώτων έως 30: {αναλογία_30:.2f} ({int(αναλογία_30 * 100)}%)")
    print(f"Πυκνότητα πρώτων έως 100: {αναλογία_100:.2f} ({int(αναλογία_100 * 100)}%)")

    print("\nΣυμπέρασμα:")
    print("Οι πρώτοι αριθμοί δεν σταματούν ποτέ, αλλά εμφανίζονται ολοένα και πιο αραιά όσο μεγαλώνει το πλήθος των φυσικών αριθμών.")
```
